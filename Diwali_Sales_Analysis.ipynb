### State

```python
df.columns
```

```python
df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
```

```python
# total numbers of orders from top 10 states

sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State', y='Orders', data= sales_state)
```

```python
#total amount/sales from top 10 states

sales_amt=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.barplot(x='State', y='Amount', data=sales_amt)
```

###### From above graphs we can see that  most of the orders are from UP, Maharashtra and karnataka respectively.

#### Marital Status

```python
sns.countplot(x='Marital_Status', data=df)
sns.set(rc={'figure.figsize':(7,5)})
```

##### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})

ax=sns.countplot(x='Occupation', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data=sales_occ)
```

##### From above grpahs we can see that most of the buyers are working in IT , Healthcare and Aviation sector
